```python
import pprint

import pandas as pd
import requests
import time
import random
import xlwt
import xlutils.copy
import xlrd
import json
import csv

def start(page,score):
   url = 'https://club.jd.com/comment/productPageComments.action'
   params = {
      "productId": "10623728509",
      "score": score,
      "sortType": "5",
      "page": page,
      "pageSize": "10",
      "isShadowSku": "0",
      "fold": "1",
      "rid": "0"
   }
   headers = {
      "user-agent": "PostmanRuntime/7.32.2",
      "content-type": "application/json;charset=gbk"
   }
   response = requests.get(url=url, params=params, headers=headers)
   print(response.json())
   return json.loads(response.text)

def parse(data):
   items = data['comments']
   for i in items:
      yield (
         i['id'],  # 用户id
         i['nickname'],  # 用户名字
         i['creationTime'],  # 时间
         i['productColor'], # 款式
         i['productSize'], # 尺码
         i['score'], #评分
         i['referenceName'],
         i['content'], # 内容
         # i['content_type']
      )


def excel(items):
   # 第一次写入
   newTable = "test.xls"  # 创建文件
   wb = xlwt.Workbook("encoding='utf-8")

   ws = wb.add_sheet('sheet1')  # 创建表
   headDate = ['id', 'nickname', 'creationTime','productColor','尺码','score','referenceName','content','content_type']  # 定义标题
   for i in range(0, 9):  # for循环遍历写入
      ws.write(0, i, headDate[i], xlwt.easyxf('font: bold on'))

   index = 1  # 行数

   for data in items:  # items是十条数据 data是其中一条（一条下有三个内容）
      for i in range(0, 9):  # 列数
         if(i<8):
            print(data[i])
            ws.write(index, i, data[i])  # 行 列 数据（一条一条自己写入）
         else:
            if(data[5]==5):
               ws.write(index,i,'pos')
            else:
               ws.write(index, i, 'neg')
      print('______________________')
      index += 1  # 等上一行写完了 在继续追加行数
      wb.save(newTable)


def another(items, j):  # 如果不是第一次写入 以后的就是追加数据了 需要另一个函数

   index = (j - 1) * 10 + 1  # 这里是 每次写入都从11 21 31..等开始 所以我才传入数据 代表着从哪里开始写入

   data = xlrd.open_workbook('test.xls')
   ws = xlutils.copy.copy(data)
   # 进入表
   table = ws.get_sheet(0)

   for test in items:
      for i in range(0, 9):  # 跟excel同理
         if(i<8):
            print(test[i])
            table.write(index, i, test[i])  # 只要分配好 自己塞入
         else:
            if (test[5] == 5):
               table.write(index, i, 'pos')
            else:
               table.write(index, i, 'neg')
      print('_______________________')

      index += 1
      ws.save('test.xls')


def main():
   j = 1  # 页面数
   judge = True  # 判断写入是否为第一次写入
   score=0
   for i in range(0, 100):
      time.sleep(random.random()+1)
      # 记得time反爬 其实我在爬取的时候没有使用代理ip也没给我封 不过就当这是个习惯吧
      first = start(j,score)
      test = parse(first)

      if judge:
         excel(test)
         judge = False
      else:
         another(test, j)
      print('第' + str(j) + '页抓取完毕\n')
      j = j + 1
      score=score+1
      if(score==4):
         score=0
   ex = pd.read_excel('test.xls')
   ex.dropna(how='all', inplace=True)
   ex.to_csv('new2.csv',encoding='utf-8')


if __name__ == '__main__':
   main()
   # 这个代码仅为全部数据下的评论而已 中差评等需要修改score！
```
